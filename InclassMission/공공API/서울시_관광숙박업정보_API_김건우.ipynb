```python
import requests
import json
import pandas as pd
```

```python
# 발급받은 api key
api_key = '656f4c6a75726c613931524b626866'
```

```python
# 사용할 api url 설정
# 서울시 공공데이터 사용 -> url에 params 값 직접 사용함
# >> 공공데이터처럼 일반 url만 쓰니 response=500 에러 발생
api_url = f'http://openapi.seoul.go.kr:8088/{api_key}/json/LOCALDATA_031101/1/5/'
```

```python
# 요청 파라미터 설정

params = {
  'KEY' : api_key,
  'TYPE' : 'json',
  'SERVICE' : 'LOCALDATA_031101',
  'START_INDEX' : '1',
  'END_INDEX' : '5'
}
```

```python
# API 요청 보내기
response = requests.get(api_url, params=params)
```

```python
# 응답 데이터 확인
response
# 200 : 정상 수신
```

```python
data = response.json()
# json(key, value 형태 파일 >> dictionary 변경)
print(json.dumps(data, indent=4, ensure_ascii=False))
# indent = 4칸 들여쓰기 사용
# ensure_ascii=False : ASCII 문자 비사용
```

```python
# 응답 데이터 확인

if response.status_code == 200:
    try:
        data = response.json()
        # json(key, value 형태 파일 >> dictionary 변경)
        print(json.dumps(data, indent=4, ensure_ascii=False))
        # indent = 4칸 들여쓰기 사용
        # ensure_ascii=False : ASCII 문자 비사용
        
        # json 데이터에서 필요한 필드만 추출 >> 리스트 저장
        items = data['LOCALDATA_031101']['row']
        parsed_data = []
        
        for item in items:
            parsed_data.append({
                'MGTNO' : item['MGTNO'],
                'BPLCNM': item['BPLCNM'],
                'APVPERMYMD': item['APVPERMYMD'],
                'APVCANCELYMD': item['APVCANCELYMD'],
                'TRDSTATENM': item['TRDSTATENM'],
                'DTLSTATENM': item['DTLSTATENM'],
                'SITETEL': item['SITETEL'],
                'RDNWHLADDR': item['RDNWHLADDR'],
                'UPDATEDT': item['UPDATEDT'],
                'UPTAENM': item['UPTAENM'],
                'REGNSENM': item['REGNSENM']
            })
            
        # 추출한 데이터를 판다스 데이터 프레임으로 변환
        df = pd.DataFrame(parsed_data)
        print(df)
        
    except json.JSONDecodeError:
        print("Error 발생")
```
