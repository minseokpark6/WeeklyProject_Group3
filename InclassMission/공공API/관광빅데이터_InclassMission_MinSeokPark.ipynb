```python
# 사용할 api url 설정
api_url = 'http://apis.data.go.kr/B551011/DataLabService/metcoRegnVisitrDDList'
```

```python
# 요청 파라미터 설정 

params = {
    'numOfRows' : '10',
    'pageNo' : '1',
    'MobileOS' : 'ETC',
    'MobileApp' : 'AppTest',
    'serviceKey' : 'api_key',
    '_type' : 'Json',
    'startYmd' : '20240313',
    'endYmd' : '20240313'
}
```

```python
# API 요청 보내기 

response = requests.get(api_url, params=params)
```

```python
# 응답 데이터 확인하기 

if response.status_code == 200:
    try:
        data = response.json()
        # print(json.dumps(data, indent= 4, ensure_ascii=False))
        
        # json 데이터에서 필요한 필드만 추출 >> 리스트 저장 
        items = data['response']['body']['items']['item']
        parsed_data = []
        
        for item in items:
            parsed_data.append({
                'areaCode' : item['areaCode'],
                'areaNm' : item['areaNm'],
                'daywkDivCd' : item['daywkDivCd'],
                'daywkDivNm' : item['daywkDivNm'],
                'touDivCd' : item['touDivCd'],
                'touDivNm' : item['touDivNm'],
                'touNum' : item['touNum'],
                'baseYmd' : item['baseYmd']
            })
            
        # 추출한 데이터를 판다스 데이터프레임으로 변환 
        df = pd.DataFrame(parsed_data)
        print(df)
        
    except json.JSONDecodeError:
        print('Error 발생')
```
