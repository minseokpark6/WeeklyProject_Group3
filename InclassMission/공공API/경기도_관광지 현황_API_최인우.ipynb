```python
import requests 
import json 
import pandas as pd 
```

```python
# 발급받은 api key 입력하세요. 

api_key = '입력받은 key 입력하세요'
```

```python
# 사용할 api url 설정
api_url = 'https://openapi.gg.go.kr/TouristResort'
```

```python
# 요청 파라미터 설정합니다. 
params = {
  'KEY' : api_key,
  'Type' :'json',
  'pIndex' : '1', 
  'pSize': '20'
}
```

```python
# API 요청 보내기
response = requests.get(api_url, params=params)
```

```python
data = response.json()
# json(key, value 형태 파일 >> dictionary 변경)로 parsing 된 데이터 
print(json.dumps(data, indent=4, ensure_ascii=False))
# indent=4 : 4칸 들여쓰기 사용 
# ensure_ascii=False : ASCII 문자 비사용 (한글이 제대로 출력)
```

```python
# 응답 데이터 확인하기 

if response.status_code == 200: 
  try: 
    data = response.json()
    # json(key, value 형태 파일 >> dictionary 변경)로 parsing 된 데이터 
    print(json.dumps(data, indent=4, ensure_ascii=False))
    # indent=4 : 4칸 들여쓰기 사용 
    # ensure_ascii=False : ASCII 문자 비사용 (한글이 제대로 출력)
    
    # json 데이터에서 필요한 필드만 추출 >> 리스트 저장 
    items = data['TouristResort'][1]['row']
    parsed_data = []
    
    for item in items:
      parsed_data.append({
        '시' : item['SIGUN_NM'],
        '관광지':item['OPERT_MAINBD_NM'],
        '위도': item['REFINE_WGS84_LOGT'],
        '경도': item['REFINE_WGS84_LAT'],
        '주소': item['REFINE_LOTNO_ADDR'],
      })
      
    # 추출한 데이터를 판다스 데이터프레임으로 변환
    df = pd.DataFrame(parsed_data)
    print(df)
    
  except json.JSONDecodeError:
    print("Error 발생")
```
