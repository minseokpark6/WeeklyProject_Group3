```python
# 박스플롯 그려보기 
def price_boxplot_temp(data, column):
    plt.figure(figsize=(8, 6))
    ax = sns.boxplot(y=data[column])

    statistics = {
        'min': data[column].min(),
        'q1': data[column].quantile(0.25),
        'median': data[column].median(),
        'q3': data[column].quantile(0.75),
        'max': data[column].max()
    }

    for key, value in statistics.items():
        ax.text(0.4 if key in ['min', 'q1', 'median'] else 0.6, value, f'{key.capitalize()}: {value}', color='blue' if key in ['min', 'q1', 'median'] else 'red', fontsize=12, ha='center')

    plt.show()

# 함수 사용 
price_boxplot_temp(df_lodging, '가격')
```

```python
def count_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    num_outliers = outliers.shape[0]
    
    return num_outliers

outlier_count = count_outliers(df_lodging, '가격')
print(f"이상치 개수: {outlier_count}")
```

2. Private room in rental unit 유형별 평균가격대비 총평점

```python
# 함수 사용
def create_scatter_plot(data, accommodation_type, x_column, y_column, title):
    df_accommodation_type = data[data['숙소유형'] == accommodation_type]
    df_temp_score_price = df_accommodation_type.groupby(x_column)[y_column].mean()
    plt.scatter(df_temp_score_price.index, df_temp_score_price)
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.title(accommodation_type + ' 유형 ' + x_column + '별 ' + y_column + ' 평균')
    plt.show()

# 함수 사용
accommodation_type = 'Private room in rental unit'
create_scatter_plot(df_lodging, accommodation_type, '총평점', '가격', f'{accommodation_type} 유형 총평점별 가격평균')
```
