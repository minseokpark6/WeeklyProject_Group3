```python
# 결과 막대그래프로 시각화
# 함수로 만들기

def barplot_show(df, column_name):
    
    # 'column_name'의 값 분포 계산
    value_counts = df[column_name].value_counts()

    # 막대그래프 생성
    plt.figure(figsize=(10, 6))
    plt.bar(value_counts.index, value_counts.values, color='skyblue')
    plt.xlabel(column_name)
    plt.ylabel('빈도수')
    plt.title(f'{column_name} 분포')
    plt.xticks(rotation=45) 
    plt.show()

barplot_show(df, '답변_평균시간')
```

```python
# 박스플롯으로 이상치 확인

def boxplot_show(df, column_name, title=None):

    # 박스플롯 생성
    plt.figure(figsize=(10, 6))
    plt.boxplot(df[column_name].dropna(), vert=False, patch_artist=True)  # 이상치를 포함한 컬럼 데이터
    # 아직 결측치 처리를 안해서 dropna()로 임의 제거 후 이상치만 확인
    plt.xlabel('값')
    plt.title(title if title else f'{column_name}의 박스플롯')
    plt.grid(True)
    plt.show()

boxplot_show(df, '욕실수', title='욕실수 박스플롯')
```

그럼 왜 리뷰 세부 항목의 null개수가 다른가?
 - 확인 필요

```python
# 데이터 확인
# null값이 상대적으로 많은 숙소_가격_리뷰점수가 null값이면서 리뷰점수가 null이 아닌 값만 추출
# 추출값에 리뷰 세부 항목들 추가해서 비교
df_temp = dat[
    (dat['review_scores_value'].isnull()) & 
    (~dat['review_scores_rating'].isnull())][['listing_url','review_scores_rating','review_scores_accuracy','review_scores_cleanliness', 'review_scores_checkin','review_scores_communication','review_scores_location','review_scores_value']]

df_temp
```

확인결과 :
