욕실수, 침실수, 침대수 KNN 대체법

```python
from sklearn.impute import KNNImputer
import pandas as pd

# 숙소 유형별로 데이터 분리
dfs = {}
for room_type in ['Entire home/apt', 'Private room', 'Shared room', 'Hotel room']:
    dfs[room_type] = df[df['숙소_유형'] == room_type]

# KNN 대체 실행
imputer = KNNImputer(n_neighbors=5)  # k 값 늘렸다가 내일 결과 나올뻔함... 조심
for room_type, data in dfs.items():
    data_imputed = data.copy()
    data_imputed[['욕실수', '침실수', '침대수']] = imputer.fit_transform(data[['욕실수', '침실수', '침대수']])
    dfs[room_type] = data_imputed

# 데이터 통합
df_imputed = pd.concat(dfs.values())
df_imputed.head()
```

```python
df = df_imputed.copy()
df
```

숙소_가격 KNN 대체법

```python
# '$' 기호와 ',' 제거 후 실수(float) 형태로 변환
df['숙소_가격'] = df['숙소_가격'].replace('[$,]', '', regex=True).astype(float)
df['숙소_가격']
```

```python
from sklearn.impute import KNNImputer
import pandas as pd
import numpy as np

# 데이터 로드 및 준비
# 여기서는 df가 이미 정의되어 있다고 가정합니다.
# df = pd.read_csv('your_data.csv')

# KNNImputer 인스턴스 생성
imputer = KNNImputer(n_neighbors=5)  # K=5 설정

# 숙소 유형별로 루프
for room_type in df['숙소_유형'].unique():
    # 숙소 유형별로 데이터 필터링
    mask = df['숙소_유형'] == room_type
    subset = df.loc[mask, ['숙소_가격']]
    
    # 결측치가 있는 경우에만 KNNImputer 적용
    if subset['숙소_가격'].isnull().any():
        # KNN 대체 실행
        imputed_data = imputer.fit_transform(subset)
        # 대체된 결과를 원래 데이터프레임에 적용
        df.loc[mask, '숙소_가격'] = imputed_data

# 결과 확인
print(df.head())
```

```python
# 나머지 결측치들은 데이터량이 적으므로 제거해도 되지만, 그 외 값들이 필요할 수 있을 것 같음.
# >> 그래서 0 값으로 대체
df.fillna(0, inplace=True)
df.isnull().sum()
```
