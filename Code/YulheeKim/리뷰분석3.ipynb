```python
import pandas as pd
from textblob import TextBlob
from googletrans import Translator, LANGUAGES

# 데이터 로드 및 샘플링
df = pd.read_csv(path)  # 실제 파일 경로와 파일명 확인
sampled_df = df.sample(n=20000, random_state=42)
data = sampled_df['comments'].astype(str)  # 'comments'는 실제 리뷰 데이터가 있는 컬럼명으로 변경

# Translator 인스턴스 생성
translator = Translator()

# 감성 분석 결과를 저장할 리스트 초기화
sentiments = []

# 감성 분석 및 번역 함수 정의
def analyze_sentiment(text):
    try:
        # TextBlob으로 감성 분석
        return TextBlob(text).sentiment.polarity
    except:
        return None

# 리뷰 번역 및 감성 분석 실행
for review in data:
    try:
        # 프랑스어 리뷰를 영어로 번역 시도
        translated_review = translator.translate(review, src='fr', dest='en').text
        # 번역된 리뷰의 감성 분석
        sentiment = analyze_sentiment(translated_review)
        sentiments.append(sentiment)
    except Exception as e:
        print(f"Translation failed for review: {review[:30]}... Error: {e}")
        sentiments.append(None)  # 번역 실패 시 None 추가

# 결과 DataFrame 생성
result_df = pd.DataFrame(sentiments, columns=['Sentiment'])

# 결과 출력
print(result_df.head())

# 결과를 CSV 파일로 저장
result_df.to_csv('sentiment_analysis_results.csv', index=False)
```

```python
import pandas as pd
from textblob import TextBlob
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter


# 랜덤으로 샘플 추출
sampled_df = merged_df_outer.sample(n=20000, random_state=1)

# 결측치 제거
sampled_df.dropna(subset=['comments'], inplace=True)

# 감성 분석 결과를 저장할 리스트
sentiments = []

# 감성 분석 수행
for comment in sampled_df['comments']:
    blob = TextBlob(comment)
    sentiment_polarity = blob.sentiment.polarity
    if sentiment_polarity > 0:
        sentiments.append('Positive')
    elif sentiment_polarity < 0:
        sentiments.append('Negative')
    else:
        sentiments.append('Neutral')

sampled_df['Sentiment'] = sentiments

# 불용어 설정
stop_words = set(stopwords.words('english'))

# 각 감성별 단어 추출
words_by_sentiment = {'Positive': [], 'Negative': [], 'Neutral': []}
for sentiment in ['Positive', 'Negative', 'Neutral']:
    text = ' '.join(sampled_df[sampled_df['Sentiment'] == sentiment]['comments'].astype(str)).lower()
    tokens = word_tokenize(text)
    filtered_tokens = [word for word in tokens if word.isalpha() and word not in stop_words]
    words_by_sentiment[sentiment] = filtered_tokens

# 각 감성별 가장 빈번한 10개 단어 추출
most_common_words_by_sentiment = {}
for sentiment, words in words_by_sentiment.items():
    word_freq = Counter(words)
    most_common_words_by_sentiment[sentiment] = word_freq.most_common(10)

# 결과 출력
for sentiment, common_words in most_common_words_by_sentiment.items():
    print(f"{sentiment} Most Common Words:")
    print(common_words)
    print('\n')
```

주제모델링
주제 모델링을 통해 리뷰 텍스트에서 자주 등장하는 주제들을 추출할 수 있습니다. 이 방법은 숙소의 특징이나 장점, 단점 등을 파악하는 데 유용합니다. 가장 널리 사용되는 알고리즘 중 하나는 LDA (Latent Dirichlet Allocation)입니다.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation


# 리뷰에서 결측치 제거
data = sampled_df['comments'].dropna()

# 단어 카운트 벡터화
tf_vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
tf = tf_vectorizer.fit_transform(data)

# LDA 모델 학습
lda = LatentDirichletAllocation(n_components=10, random_state=0)
lda.fit(tf)

# 주제 출력 함수
def print_topics(model, vectorizer, top_n=10):
    for idx, topic in enumerate(model.components_):
        print("Topic %d:" % (idx))
        print([(vectorizer.get_feature_names_out()[i], topic[i])
               for i in topic.argsort()[:-top_n - 1:-1]])

print_topics(lda, tf_vectorizer)
```

고빈도 키워드 추출
특정 단어가 리뷰에서 자주 등장한다면, 그 단어가 숙소의 중요한 특징을 나타낼 가능성이 높습니다. TF-IDF 방식을 사용하여 중요 단어를 식별할 수 있습니다.

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from collections import Counter

# 데이터 준비
data = sampled_df['comments'].dropna()
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(data)

# 특성 이름 가져오기
feature_array = np.array(tfidf_vectorizer.get_feature_names_out())

# 각 문서에 대해 상위 TF-IDF 값을 가진 단어 인덱스 찾기
top_n = 20
sorted_indices = np.argsort(tfidf_matrix.toarray(), axis=1)[:, -top_n:]

# 상위 특성 추출
top_features = [feature_array[indices] for indices in sorted_indices]

# 전체 문서에 대해 가장 많이 등장하는 키워드 추출
all_top_features = [feature for sublist in top_features for feature in sublist]
top_feature_counts = Counter(all_top_features)

# 가장 빈번한 20개의 키워드 출력
most_common_features = top_feature_counts.most_common(20)
print(most_common_features)

```

n-gram 분석
n-gram 분석을 통해 단어들의 조합을 파악하여, 어떤 표현이 자주 사용되는지 분석할 수 있습니다. 이는 숙소의 특정 특성이나 서비스에 대한 인사이트를 제공할 수 있습니다.

```python
from sklearn.feature_extraction.text import CountVectorizer
from collections import Counter

# n-gram 범위 설정 및 불용어 처리를 위한 CountVectorizer 초기화
vectorizer = CountVectorizer(ngram_range=(2, 3), stop_words='english')
X2 = vectorizer.fit_transform(data)

# 특성 이름 가져오기
features = vectorizer.get_feature_names_out()

# 행렬을 배열로 변환 후 각 특성의 합계를 계산
counts = X2.sum(axis=0).A1

# 특성과 그 횟수를 매핑하여 빈도 분석
freq_distribution = Counter(dict(zip(features, counts)))

# 가장 흔한 n-gram 출력
print(freq_distribution.most_common(10))

```
