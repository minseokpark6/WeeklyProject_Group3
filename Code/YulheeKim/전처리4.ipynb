```python
import pandas as pd
from sklearn.impute import SimpleImputer

# 데이터 로드

df = pd.read_csv(path)

# 'price' 컬럼의 '$' 제거 및 형 변환
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)

# 데이터 확인
print(df.head())
print(df.isnull().sum())

# 'bedrooms'와 'bathrooms'를 기준으로 그룹화하고 평균으로 결측치 대체
grouped = df.groupby(['bedrooms', 'bathrooms'])
def impute_group(group):
    imputer = SimpleImputer(strategy='mean')
    group['price'] = imputer.fit_transform(group[['price']])
    return group

df_imputed = grouped.apply(impute_group)

# 결과 확인 및 저장
print(df_imputed.head())
print(df_imputed.isnull().sum())
df_imputed.to_csv('path_to_your_output.csv', index=False)

```

```python
# 데이터프레임의 각 컬럼별 결측치 수 확인
missing_values = df_imputed.isnull().sum()
print(missing_values)
```
