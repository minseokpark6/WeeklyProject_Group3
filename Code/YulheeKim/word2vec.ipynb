CBOW 벡터의 특징
CBOW 모델에서 airbnb의 벡터는 다양한 차원에 걸쳐 상당히 큰 양수와 음수 값들을 포함하고 있습니다. 이는 모델이 단어의 맥락을 기반으로 airbnb의 의미를 폭넓게 잡아내려고 시도했음을 나타냅니다.
예를 들어, 값 3.3224702는 이 차원이 airbnb와 매우 강력하게 양의 관련성을 가지고 있음을 의미할 수 있습니다. 반면, -3.3740335와 같은 음수 값은 이 차원이 airbnb의 반대 특성을 표현할 수 있습니다.
Skip-Gram 벡터의 특징
Skip-Gram 모델에서 airbnb의 벡터는 CBOW보다 상대적으로 낮은 절대값을 보이고 있습니다. 이는 Skip-Gram 모델이 특정 단어를 기반으로 주변 단어를 예측하는 방식 때문에, 개별 차원의 값들이 더 분산되고 미세 조정될 수 있음을 시사합니다.
Skip-Gram 벡터의 각 값은 airbnb 주변에 자주 등장하는 단어들과의 관계를 나타냅니다. 예를 들어, 0.7205999는 이 차원에서 airbnb가 강한 양의 연관성을 가진 특성이나 단어와 연결될 수 있음을 나타내며, -0.7749699는 강한 음의 연관성을 가리킵니다.
임베딩 벡터의 활용
이러한 벡터는 유사 단어 탐색, 클러스터링, 분류 작업, 감성 분석 등 다양한 자연어 처리 작업에 활용될 수 있습니다. 예를 들어, 벡터 간 코사인 유사도를 계산하여 airbnb와 유사한 단어를 찾거나, 여러 단어의 벡터를 평균내어 특정 주제나 문맥을 대표하는 벡터를 생성할 수 있습니다.

이 임베딩은 단순히 수치로 보기보다는 그것들이 어떻게 계산되고 활용되는지를 이해하는 것이 중요합니다. 각 차원의 구체적인 의미를 직접적으로 해석하는 것은 어렵지만, 전체적인 벡터 구조는 단어의 문맥적, 의미적 특성을 잘 반영합니다.

```python
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from collections import Counter
import nltk
from nltk.tokenize import word_tokenize

# 필요한 NLTK 자원 다운로드
nltk.download('vader_lexicon')
nltk.download('punkt')
nltk.download('stopwords')

# 데이터 로드
data = merged_df_outer

# 랜덤으로 50,000개의 리뷰 선택
sampled_data = data.sample(n=50000, random_state=1)

# 영어와 프랑스어 불용어 목록 로드
english_stop_words = set(stopwords.words('english'))
french_stop_words = set(stopwords.words('french'))

# 언어별 불용어 제거 함수 정의
def remove_stopwords(text, language='english'):
    tokens = word_tokenize(text)
    if language == 'english':
        filtered_tokens = [word for word in tokens if word not in english_stop_words and word.isalpha()]
    else:
        filtered_tokens = [word for word in tokens if word not in french_stop_words and word.isalpha()]
    return " ".join(filtered_tokens)

# 감성 분석기 초기화
sia = SentimentIntensityAnalyzer()

# 리뷰가 영어인지 프랑스어인지를 판별하는 로직이 필요할 수 있음
# 여기서는 모든 리뷰를 영어로 가정하고 진행
# 데이터 타입을 문자열로 변환하고 불용어 제거
sampled_data['comments'] = sampled_data['comments'].astype(str).apply(lambda x: remove_stopwords(x, 'english'))

# 감성 점수 계산 및 긍정 리뷰 필터링
sampled_data['sentiment'] = sampled_data['comments'].apply(lambda x: sia.polarity_scores(x)['compound'])
positive_reviews = sampled_data[sampled_data['sentiment'] > 0.5]  # 예를 들어, compound 점수가 0.5 이상인 리뷰 선택

# 긍정 리뷰에서 자주 등장하는 단어 추출
word_counts = Counter(" ".join(positive_reviews['comments']).split())
most_common_words = word_counts.most_common(20)  # 상위 20개 단어

print(most_common_words)
```

CBOW 벡터 해석
CBOW 모델에서 'apartment'의 벡터는 일부 차원에서 매우 높은 양수값(예: 3.5883152, 4.089871)과 음수값(예: -3.9510322, -3.511913)을 포함하고 있습니다. 이런 큰 값을 보이는 차원들은 해당 단어의 문맥 내에서 중요한 의미적 특성을 캡처할 가능성이 높습니다. CBOW는 주변 단어를 통해 타겟 단어를 예측하기 때문에, 이러한 큰 값들은 주변 단어와의 강한 연관성을 나타낼 수 있습니다.

Skip-Gram 벡터 해석
Skip-Gram 모델의 벡터는 일반적으로 더 작고 균일한 값들을 포함하고 있으며, 이는 Skip-Gram 모델이 타겟 단어로부터 주변 단어를 예측하는 데 초점을 맞추기 때문입니다. 여기서 값들이 비교적 작은 것은 단어 간의 관계를 미세하게 조정하기 때문일 수 있습니다. 예를 들어, 0.5285611, -0.87354934와 같은 값들은 주변 단어와의 관계에서 중요한 특성을 나타낼 수 있습니다.

벡터의 활용
이러한 벡터들은 다양한 방식으로 활용될 수 있습니다:

유사도 측정: 다른 단어들과의 코사인 유사도를 계산하여 의미적 유사성을 파악할 수 있습니다.
클러스터링: 여러 단어의 벡터를 사용하여 클러스터링을 수행하고, 유사한 의미를 가진 단어들을 그룹화할 수 있습니다.
분류 작업: 임베딩을 특성으로 사용하여 텍스트 분류나 감성 분석과 같은 작업을 수행할 수 있습니다.
각 벡터의 차원은 특정한 '의미'를 직접적으로 해석하기 어렵지만, 전체적으로 단어의 의미적 및 문맥적 속성을 잘 반영한다고 볼 수 있습니다.

```python
import pandas as pd
from gensim.models import Word2Vec
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import numpy as np

# 필요한 NLTK 자원 다운로드
nltk.download('punkt')
nltk.download('stopwords')

# 데이터 로드 (CSV 파일을 예시로 함)
data = merged_df_outer

# 랜덤으로 50,000개의 리뷰 선택
sampled_data = data.sample(n=50000, random_state=1)['comments']  # 'review_text'는 리뷰 텍스트가 있는 컬럼 이름

# 텍스트를 소문자로 변환하고 토큰화
def preprocess_text(text):
    if pd.isna(text):
        return []  # 비어 있는 경우 빈 리스트 반환
    stop_words = set(stopwords.words('english'))  # 언어에 따라 'french' 등을 추가
    tokens = word_tokenize(text.lower())
    return [word for word in tokens if word.isalpha() and word not in stop_words]

# 전처리 수행
processed_data = sampled_data.apply(preprocess_text).tolist()

# 모델 초기화 및 어휘 구축
model = Word2Vec(vector_size=100, window=5, min_count=5, workers=4)
model.build_vocab(processed_data)  # 어휘 구축

# 모델 훈련
model.train(processed_data, total_examples=model.corpus_count, epochs=model.epochs)

# 'good', 'best'와 같은 형용사 주변에서 숙소와 관련된 키워드 찾기
keywords = ['good', 'best', 'comfortable', 'great', 'nice']
related_words = set()

for keyword in keywords:
    if keyword in model.wv:
        # 각 키워드에 대해 가장 유사한 단어 10개 추출
        similar_words = model.wv.most_similar(keyword, topn=10)
        related_words.update([word for word, _ in similar_words])

# 결과 출력
print("Related words to positive adjectives:", related_words)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.manifold import TSNE
from gensim.models import Word2Vec
import pandas as pd

# 이미 학습된 Word2Vec 모델이 있다고 가정
# model = Word2Vec.load("your_model_path")

# 관심 있는 단어 목록
keywords = ['good', 'best', 'comfortable', 'great', 'nice']
related_words = set(keywords)

# 관심 있는 단어와 가장 유사한 단어들을 찾기
for keyword in keywords:
    if keyword in model.wv:
        similar_words = model.wv.most_similar(keyword, topn=10)
        related_words.update([word for word, _ in similar_words])

# 유사한 단어들의 벡터 추출
word_vectors = np.array([model.wv[word] for word in related_words if word in model.wv])

# t-SNE로 차원 축소
tsne = TSNE(n_components=2, random_state=0)
word_vectors_2d = tsne.fit_transform(word_vectors)

# 그래프로 시각화
plt.figure(figsize=(10, 10))
for word, coord in zip(related_words, word_vectors_2d):
    x, y = coord
    plt.scatter(x, y)
    plt.text(x, y, word, fontsize=9)

plt.title('Word2Vec Word Embeddings Visualized with t-SNE')
plt.xlabel('t-SNE feature 0')
plt.ylabel('t-SNE feature 1')
plt.grid(True)
plt.show()
```

```python
from gensim.models import Word2Vec

# 예를 들어 모델이 이미 학습되었다고 가정
# model = Word2Vec.load("your_model_path")

# 분석할 단어 리스트
words = {'liked', 'fantastic', 'tidy', 'fabulous', 'reasonable', 'bunk', 'lovely', 'convenient', 'good', 
         'neat', 'brilliant', 'also', 'sparkling', 'decent', 'favorite', 'favourite', 'incredible', 'terrific',
         'nicest', 'pleasant', 'superb', 'spacious', 'perfect', 'nice', 'firm', 'amazing', 'cute', 'comfy', 
         'great', 'spotless', 'wonderful', 'awesome', 'cool', 'cozy'}

# 각 단어와 'hotel'의 코사인 유사도 계산 및 가장 유사한 단어 찾기
similarities = {}
top_similar_words = {}

for word in words:
    if word in model.wv:  # 모델의 어휘 사전에 단어가 있는지 확인
        # 'hotel'과의 유사도
        similarity = model.wv.similarity('hotel', word)
        similarities[word] = similarity

        # 해당 단어와 유사한 상위 5개 단어 찾기
        similar_words = model.wv.most_similar(word, topn=5)
        top_similar_words[word] = similar_words

# 유사도 결과 출력
print("Cosine similarities with 'hotel':")
for word, sim in sorted(similarities.items(), key=lambda item: item[1], reverse=True):
    print(f"{word}: {sim:.3f}")

# 유사한 단어 결과 출력
print("\nTop similar words for each word in the list:")
for word, sim_words in top_similar_words.items():
    print(f"{word}: {', '.join([w for w, _ in sim_words])}")
```
