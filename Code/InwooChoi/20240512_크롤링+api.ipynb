```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time
import pandas as pd
from bs4 import BeautifulSoup

# Chrome Driver ì„¤ì •
chrome_options = Options()
chrome_options.add_argument("--headless")  # ë°±ê·¸ë¼ìš´ë“œì—ì„œ ì‹¤í–‰
chrome_options.add_argument("--start-maximized")  # ìµœëŒ€í™”ëœ ìƒíƒœë¡œ ì‹œì‘
chrome_options.add_experimental_option("detach", True)  # ìŠ¤í¬ë¦½íŠ¸ ì¢…ë£Œ í›„ì—ë„ ë¸Œë¼ìš°ì € ìœ ì§€
chrome_options.add_experimental_option("excludeSwitches", ["enable-logging"])  # ë¶ˆí•„ìš”í•œ ë¡œê·¸ ì œì™¸
chrome_service = Service('C:/chromedriver-win64/chromedriver.exe')  # í¬ë¡¬ ë“œë¼ì´ë²„ ê²½ë¡œ

def get_listing_details(url):
    driver = webdriver.Chrome(service=chrome_service, options=chrome_options)
    driver.get(url)
    
    # í˜ì´ì§€ ë¡œë“œ ëŒ€ê¸°
    wait = WebDriverWait(driver, 2)  # ëŒ€ê¸° ì‹œê°„ì„ 30ì´ˆë¡œ ì¦ê°€
    try:
        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, 'div[data-testid="property-card"]')))
    except TimeoutException:
        print("í˜ì´ì§€ ë¡œë”© ì‹œê°„ ì´ˆê³¼")
        driver.quit()
        return []

    listings = []
    
    while True:
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        listing_elements = soup.find_all('div', {'data-testid': 'property-card'})
        print(f'Found {len(listing_elements)} listing elements.')

        for listing in listing_elements:
            try:
                # ìˆ™ì†Œ ìƒì„¸ í˜ì´ì§€ë¡œ ì´ë™
                listing_link = listing.find('a', {'data-testid': 'title-link'})['href']
                driver.get(f"https://www.airbnb.co.kr{listing_link}")
                time.sleep(5)  # í˜ì´ì§€ ë¡œë”© ì‹œê°„ ëŒ€ê¸°

                # ìˆ™ì†Œ ìƒì„¸ ì •ë³´ ì¶”ì¶œ
                soup = BeautifulSoup(driver.page_source, 'html.parser')
                room_name = soup.find('h1', {'data-testid': 'title'}).text.strip()
                review_score = soup.find('span', {'aria-label': lambda text: text and "ë³„ì " in text}).text.strip()
                reviews = []
                
                # ëª¨ë“  ë¦¬ë·° í˜ì´ì§€ íƒìƒ‰
                while True:
                    review_elements = soup.find_all('div', {'data-review-id': True})

                    for element in review_elements:
                        review_text = element.find('span', {'class': '_1epc9dn0'}).text
                        reviews.append(review_text)

                    # ë‹¤ìŒ ë¦¬ë·° í˜ì´ì§€ë¡œ ì´ë™
                    try:
                        next_button = driver.find_element(By.CSS_SELECTOR, 'button[aria-label="ë‹¤ìŒ í˜ì´ì§€"]')
                        next_button.click()
                        time.sleep(5)  # í˜ì´ì§€ ë¡œë”© ì‹œê°„ ëŒ€ê¸°
                        soup = BeautifulSoup(driver.page_source, 'html.parser')
                    except:
                        break  # ë” ì´ìƒ ë‹¤ìŒ í˜ì´ì§€ê°€ ì—†ìœ¼ë©´ ì¢…ë£Œ

                listings.append({
                    'room_name': room_name,
                    'review_score': review_score,
                    'reviews': reviews
                })

                # ë’¤ë¡œ ê°€ê¸°
                driver.back()
                time.sleep(5)  # í˜ì´ì§€ ë¡œë”© ì‹œê°„ ëŒ€ê¸°
                soup = BeautifulSoup(driver.page_source, 'html.parser')
            except Exception as e:
                print(f"Error: {e}")
                continue

        # ë‹¤ìŒ í˜ì´ì§€ë¡œ ì´ë™
        try:
            next_button = driver.find_element(By.CSS_SELECTOR, 'a[aria-label="ë‹¤ìŒ í˜ì´ì§€"]')
            next_button.click()
            time.sleep(5)  # í˜ì´ì§€ ë¡œë”© ì‹œê°„ ëŒ€ê¸°
        except:
            break  # ë” ì´ìƒ ë‹¤ìŒ í˜ì´ì§€ê°€ ì—†ìœ¼ë©´ ì¢…ë£Œ

    driver.quit()
    return listings

def save_listings_to_csv(url, output_file):
    listings = get_listing_details(url)
    if listings:
        df = pd.DataFrame(listings)
        df.to_csv(output_file, index=False)
        print(f"Saved listings to {output_file}")
    else:
        print("No listings found.")

# LA ìˆ™ì†Œ ëª©ë¡ URL
url = 'https://www.airbnb.co.kr/s/%EB%89%B4%EC%9A%95/homes?place_id=ChIJOwg_06VPwokRYv534QaPC8g&refinement_paths%5B%5D=%2Fhomes&checkin=2024-05-13&checkout=2024-05-15&date_picker_type=calendar&adults=0&children=0&infants=0&pets=0&search_type=AUTOSUGGEST'
save_listings_to_csv(url, 'airbnb_la_listings.csv')
```
