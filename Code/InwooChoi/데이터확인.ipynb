```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from concurrent.futures import ThreadPoolExecutor
import time  # time 모듈 import 추가

# ChromeDriver 경로 설정
chrome_driver_path = 'C:/chromedriver-win64/chromedriver.exe'  # 실제 chromedriver 경로로 변경
chrome_service = Service(chrome_driver_path)

# Selenium 옵션 설정
options = Options()
#options.add_argument("--headless")  # 브라우저를 띄우지 않음 (필요에 따라 주석 처리)
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")


def crawl_airbnb_details(url):
    # Selenium WebDriver 초기화
    driver = webdriver.Chrome(service=chrome_service, options=options)

    # 브라우저 창 크기 설정
    driver.set_window_size(1920, 1080)  # 예: 1920x1080 해상도
    driver.get(url)
    time.sleep(1)  # 페이지가 로딩될 시간을 충분히 주기

    # 2 guestsStudio 1 bed 1 bath 부분 가져오기
    try:
        details_section = WebDriverWait(driver, 2).until(
            EC.presence_of_element_located((By.XPATH, '//ol[contains(@class, "lgx66tx atm")]'))
        )
        details_text = details_section.text if details_section else 'Details not found'
    except Exception as e:
        details_text = f'Details not found: {e}'

    # Response rate 부분 가져오기 (JavaScript 사용)
    try:
        response_rate_text = driver.execute_script("""
            return document.querySelector("#site-content > div > div:nth-child(1) > div:nth-child(6) > div > div > div > div > section > div.cmuek5n.atm_2d_1ro4147.atm_lo_1vi7ecw.atm_le_1tcgj5g.atm_lo_1ylpe5n__oggzyc.c8clt7m.atm_ll_1tcgj5g.atm_lk_1tcgj5g.atm_9s_1txwivl.atm_fc_1h6ojuz.c1wueegw.atm_5j_qe0vi4__oggzyc.dir.dir-ltr > div > div > div:nth-child(2) > div:nth-child(2) > div > div:nth-child(2) > div:nth-child(1)").innerText;
        """)
        if not response_rate_text:
            response_rate_text = 'Response rate not found'
    except Exception as e:
        response_rate_text = f'Response rate not found: {e}'

    # Responds within a few hours 부분 가져오기 (JavaScript 사용)
    try:
        response_time_text = driver.execute_script("""
            return document.querySelector("#site-content > div > div:nth-child(1) > div:nth-child(6) > div > div > div > div > section > div.cmuek5n.atm_2d_1ro4147.atm_lo_1vi7ecw.atm_le_1tcgj5g.atm_lo_1ylpe5n__oggzyc.c8clt7m.atm_ll_1tcgj5g.atm_lk_1tcgj5g.atm_9s_1txwivl.atm_fc_1h6ojuz.c1wueegw.atm_5j_qe0vi4__oggzyc.dir.dir-ltr > div > div > div:nth-child(2) > div:nth-child(2) > div > div:nth-child(2) > div:nth-child(2)").innerText;
        """)
        if not response_time_text:
            response_time_text = 'Response time not found'
    except Exception as e:
        response_time_text = f'Response time not found: {e}'

    driver.quit()
    return details_text, response_rate_text, response_time_text

# 병렬 처리할 스레드 수
num_threads = 4  # 예: 4개의 스레드로 병렬 처리

# 크롤링 결과를 저장할 리스트
results = []

# 각 URL에 대해 크롤링을 수행하고 결과를 리스트에 저장
with ThreadPoolExecutor(max_workers=num_threads) as executor:
    futures = [executor.submit(crawl_airbnb_details, row['url']) for index, row in df_with_nulls.iloc[:100].iterrows()]

    # 각 크롤링 작업이 완료되면 결과를 리스트에 저장
    for future in futures:
        results.append(future.result())

# 결과를 데이터프레임으로 변환
df_results = pd.DataFrame(results)

# 결과 출력
print(df_results)
```
