```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup


# List of URLs to scrape
lodging_urls = df[(df['욕실수'].isna()) & (df['침실수'].isna()) & (df['침대수'].isna())][['숙소_id','숙소_url']].values

# Setup Chrome options
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)
chrome_options.add_experimental_option("excludeSwitches", ['enable-logging'])

# Initialize the Chrome driver
driver = webdriver.Chrome(options=chrome_options)
driver.set_window_size(1920, 1080)  # Set browser window size

# Function to scrape data from a given URL
def scrape_airbnb(url_info):
    host_id, url = url_info
    driver.get(url)
    time.sleep(3)  # Allow time for the page to load

    # Close the translation popup if it appears
    # try:
    #     close_button = driver.find_element(By.CSS_SELECTOR, "body > div:nth-child(50) > div > div > section > div > div > div.p1psejvv.atm_9s_1bgihbq.dir.dir-ltr > div > div.c1lbtiq8.atm_mk_stnw88.atm_9s_1txwivl.atm_fq_1tcgj5g.atm_wq_kb7nvz.atm_tk_1tcgj5g.dir.dir-ltr > button")
    #     close_button.send_keys(Keys.ENTER)
    # except Exception as e:
    #     print(f"No translation popup to close for {url}")

    # Parse the page content
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    # Extract the number of bedrooms, beds, and bathrooms
    try:
        etc = driver.find_element(By.XPATH, '//ol[contains(@class, "lgx66tx atm")]')
        etc_text = etc.text
        return host_id, url, etc_text
    except Exception as e:
        etc_text = "N/A"
        print(f"Could not find the info for {url}: {e}")
        return host_id, url, etc_text

# Loop through each URL and scrape data
results = []
for url_info in lodging_urls[:1000]:  # Slice the list to iterate over the first 1000 entries
    # print(f"Scraping URL: {url_info[1]}")
    result = scrape_airbnb(url_info)
    results.append(result)
    time.sleep(2)  # Optional: pause between requests

# Close the browser after scraping
driver.quit()

# Create DataFrame from results
results_df = pd.DataFrame(results, columns=['호스트_id', '숙소_url', 'etc'])

# Save DataFrame to CSV
results_df.to_csv('scraped_data.csv', index=False)

print("Data saved to 'scraped_data_1.csv'")
```
