```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup


# 욕실, 침실, 침대수가 nan값인 숙소_id, 숙소_url 데이터의 값 7806개 
lodging_urls = df[(df['욕실수'].isna()) & (df['침실수'].isna()) & (df['침대수'].isna())][['숙소_id','숙소_url']].values

#크롬 option 설정 꼭 입력!
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)
chrome_options.add_experimental_option("excludeSwitches", ['enable-logging'])
driver = webdriver.Chrome(options=chrome_options)

#웹 브라우저 사이즈 고장
driver.set_window_size(1920, 1080)  


def scrape_airbnb(idnex, url_info):
    host_id, url = url_info
    driver.get(url)
    time.sleep(2) #로딩될 시간 지정 2초 

    

    # bs4 설정 
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    # 침대, 침실, 욕실수 크롤링
    try:
        etc = driver.find_element(By.XPATH, '//ol[contains(@class, "lgx66tx atm")]')
        etc_text = etc.text
        return host_id, url, etc_text
    except Exception as e:
        etc_text = "N/A"
        print(f"Could not find the info for {url}: {e}")
        return host_id, url, etc_text

# lodging_url의 숙소_url에 맞는 침대,침실,욕실 수 불러오기 
results = []
for index, url_info in enumerate(lodging_urls[:1000]):  # 1000개만 해봄
    print(f"{index + 1} ")
    result = scrape_airbnb(index, url_info)
    results.append(result)
    time.sleep(1)  

# 다 돌리면 크롤링 끄기
driver.quit()

# 결과값 데이터프레임화 
results_df = pd.DataFrame(results, columns=['호스트_id', '숙소_url', 'etc'])

# csv파일로 저장
results_df.to_csv('scraped_data.csv', index=False)

print("Data saved to 'scraped_data_1.csv'")
```
